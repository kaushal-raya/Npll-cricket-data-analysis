# *Which teams are most efficient?*

```python
df.columns
```

```python
df2.groupby('team_1')['team_1_runs'].sum().sort_values(ascending =False)
```

```python
df2.groupby('team_2')['team_2_runs'].sum().sort_values(ascending =False)
```

```python
team_1_runs = {'Sudurpaschim Royals': 996, 'Karnali Yaks': 945, 'Chitwan Rhinos': 635, 
               'Biratnagar Kings': 564, 'Lumbini Lions': 528, 'Pokhara Avengers': 379, 
               'Kathmandu Gurkhas': 315, 'Janakpur Bolts': 275}
team_2_runs = {'Janakpur Bolts': 1024, 'Pokhara Avengers': 537, 'Chitwan Rhinos': 532, 
               'Kathmandu Gurkhas': 509, 'Lumbini Lions': 485, 'Biratnagar Kings': 411, 
               'Sudurpaschim Royals': 377, 'Karnali Yaks': 238}

team_1_df = pd.DataFrame(list(team_1_runs.items()), columns=['team', 'runs_1'])
team_2_df = pd.DataFrame(list(team_2_runs.items()), columns=['team', 'runs_2'])

merged_df = pd.merge(team_1_df, team_2_df, on='team', how='outer').fillna(0)

merged_df['total_runs'] = merged_df['runs_1'] + merged_df['runs_2']

result = merged_df[['team', 'total_runs']].sort_values(by='total_runs', ascending=False)

print(result)
```

```python
efficiency=merged_df['total_runs']/(df2['total_ball_bowled']-df2['extra_balls'])

```

```python
df2['team']
```

```python
df2['Effi']=efficiency
```

```python
a=df2.fillna(0)
```

```python
team1 = df2[['team_1', 'team_1_runs', 'total_ball_bowled', 'extra_balls']].copy()
team1.rename(columns={'team_1': 'team', 'team_1_runs': 'runs'}, inplace=True)
team1['efficiency'] = team1['runs'] / (df2['total_ball_bowled'] - df2['extra_balls'])

team2 = df2[['team_2', 'team_2_runs', 'total_ball_bowled', 'extra_balls']].copy()
team2.rename(columns={'team_2': 'team', 'team_2_runs': 'runs'}, inplace=True)
team2['efficiency'] = team2['runs'] / (df2['total_ball_bowled'] - df2['extra_balls'])

eff_df = pd.concat([team1, team2], ignore_index=True)

eff_df = eff_df.dropna()
eff_df = eff_df[eff_df['efficiency'].notnull() & (eff_df['efficiency'] != float('inf'))]

team_efficiency = eff_df.groupby('team')['efficiency'].mean().sort_values(ascending=False).reset_index()
team_efficiency.columns = ['Team', 'Average Efficiency']
team_efficiency['Average Efficiency'] = team_efficiency['Average Efficiency'].round(3)

print(team_efficiency)

plt.figure(figsize=(11, 6))
sns.barplot(data=team_efficiency, y='Average Efficiency', x='Team',hue='Average Efficiency', palette='viridis')
plt.title('Team Efficiency in NPL (Runs per Legal Delivery)')
plt.xlabel('Efficiency (Runs per Ball)')
plt.xticks(rotation=45)
plt.ylabel('Team')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.savefig('Efficiency.png')

plt.show()
```
